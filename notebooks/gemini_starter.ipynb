## Objectif

Ce notebook sert de point de départ pour interagir avec l'API Gemini via le SDK officiel `google-genai`.

Il couvre les étapes minimales pour :
1. Charger la clé API depuis un fichier `.env` local
2. Initialiser le client Gemini
3. Envoyer une première requête de génération de texte

> **Pré-requis :** Copier `.env.example` en `.env` à la racine du projet et y renseigner votre `GEMINI_API_KEY`.

```python
# Chargement de la clé API depuis .env
import os
from pathlib import Path
from dotenv import load_dotenv

# Remonte d'un niveau pour trouver .env à la racine du projet
env_path = Path("__file__").resolve().parent.parent / ".env"
load_dotenv(dotenv_path=env_path)

api_key = os.environ.get("GEMINI_API_KEY")

if not api_key:
    raise EnvironmentError(
        "GEMINI_API_KEY introuvable. "
        "Copiez .env.example en .env et renseignez votre clé."
    )

print("Clé API chargée avec succès.")
```

```python
# Initialisation du client google-genai
from google import genai

client = genai.Client(api_key=api_key)

MODEL = "gemini-2.0-flash"

print(f"Client Gemini initialisé. Modèle : {MODEL}")
```

```python
# Test simple de génération de texte
prompt = "Explique en deux phrases ce qu'est un agent IA."

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
)

print("=== Réponse Gemini ===")
print(response.text)
```
